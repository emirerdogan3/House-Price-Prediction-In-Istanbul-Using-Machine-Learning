```python
import pandas as pd
import os

file_names = [
    'buying_Prices_V3_sayfa_36_77.xlsx',
    'buying_Prices_V3_sayfa_77_114.xlsx',
    'buying_Prices_V3_sayfa_114_149.xlsx',
    'buying_Prices_V3_sayfa_149_195.xlsx',
    'buying_Prices_V3_sayfa_195_250.xlsx',
    'buying_Prices_V3_sayfa_250_261.xlsx'
]

all_data_frames = []

for file in file_names:
    try:
        df = pd.read_excel(file)

        all_data_frames.append(df)
        print(f"'{file}' başarıyla okundu.")
    except FileNotFoundError:
        print(f"UYARI: '{file}' dosyası bulunamadı. Lütfen dosya adını ve konumunu kontrol edin.")
    except Exception as e:
        print(f"'{file}' okunurken bir hata oluştu: {e}")

if all_data_frames:
    combined_df = pd.concat(all_data_frames, ignore_index=True)

    output_filename = 'birlesik_ev_fiyatlari_veriseti.xlsx'
    combined_df.to_excel(output_filename, index=False)

    print("\n----------------------------------------------------")
    print(f"Sonuçlar '{output_filename}' dosyasına kaydedildi.")
    print(f"Toplam Satır Sayısı: {len(combined_df)}")
    print(f"Toplam Sütun Sayısı: {len(combined_df.columns)}")
    print("----------------------------------------------------")
else:
    print("\nHiçbir dosya okunamadığı için birleştirme işlemi yapılamadı.")
```
